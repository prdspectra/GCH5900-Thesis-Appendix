```python
#import dependencies
import csv
import os
import pandas as pd
import numpy as np
import glob
import shutil
```

```python
#assign directories
indirectory = '/Users/pierce/Documents/Ex 3/Trimmed Data/'
outdirectory = '/Users/pierce/Documents/Ex 3/Outputs/'
```

```python
#create dataframe
df = pd.read_csv(outdirectory + 'merged_data.csv')
df
```

```python
#create new dataframes holding the relevant data
third = df.iloc[2::9, :]
fourth = df.iloc[3::9, :]
fifth = df.iloc[4::9, :]
```

```python
#concatenate the dataframes together
tdf = pd.concat([third,fourth,fifth])
tdf.sort_index(inplace = True)
tdf.reset_index(inplace = True)
tdf.drop(['index', 'Unnamed: 0'], axis = 1, inplace = True)
display(tdf)
```

```python
#define new dataframe to hold mean values
mdf = pd.DataFrame(columns = df.columns)
mdf.drop('Unnamed: 0', axis = 1, inplace = True)
display(mdf)


#slice the dataframe into triplicates and calculate the mean
for i in range(0, len(tdf), 3):
    subset = tdf.iloc[i:i+3, :]
    meanrow = subset.mean(axis = 0)
    mdf_temp = meanrow.to_frame().transpose()
    mdf.loc[len(mdf)] = mdf_temp.iloc[0,:]
    
display(mdf)
```

```python
concentration = []

for i in range(0,101,2):
    concentration.append(i)
    
mdf['MA Concentration (%)'] = concentration
display(mdf)
```

```python
mdf.to_csv(outdirectory + 'mean_data.csv', index = True)
```
